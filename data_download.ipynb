```python
import pandas as pd
import requests
import os
from urllib.parse import urlparse
```

```python
def download_image(url, folder_path):
    try:
        response = requests.get(url, stream=True)
        response.raise_for_status()
        # Extract the filename from the URL
        filename = os.path.basename(urlparse(url).path)
        filepath = os.path.join(folder_path, filename)
        with open(filepath, 'wb') as out_file:
            for chunk in response.iter_content(chunk_size=8192):
                out_file.write(chunk)
        print(f"Downloaded: {filename}")
    except requests.exceptions.HTTPError as http_err:
        print(f"HTTP error occurred: {http_err}")
    except Exception as err:
        print(f"Other error occurred: {err}")
```

```python
excel_file = 'image-url.xlsx'  
sheet_name = 'Sheet 1'  
column_name = 'Table 1'
```

```python
try:
    df = pd.read_excel(excel_file, sheet_name=sheet_name)
    print("Excel file read successfully.")
    print("Available columns:", df.columns)
except Exception as e:
    print(f"Error reading the Excel file: {e}")
    exit()
```

```python
folder_path = 'downloaded_images'
if not os.path.exists(folder_path):
    os.makedirs(folder_path)
```

```python
try:
    for index, row in df.iterrows():
        try:
            url = row[column_name]
            download_image(url, folder_path)
        except KeyError as e:
            print(f"Column '{column_name}' not found in the row: {e}")
except KeyboardInterrupt:
    print("\nDownload process interrupted by user.")

print("All images downloaded or download stopped by user!")
```
